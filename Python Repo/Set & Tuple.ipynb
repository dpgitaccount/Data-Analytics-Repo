# Set & Tuple

# SET

```python
## Creating a set

my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
print('-'*50)

# Adding
my_set.add(6)
print(my_set)
print('-'*50)

## pop
popped = my_set.pop()
print(popped)
print('-'*50)

## Remove
print(my_set)
my_set.discard(3)
print(my_set)
print('-'*50)

## Iterating
print(my_set)
for i in my_set:
    print(i)

print('-'*50)

## checking the values
print(my_set)
print(2 in my_set)
print(5 in my_set)
print(8 in my_set)

print('-'*50)

## Set operations
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9}
print('Union :', set1 | set2)
print('Intersection :', set1 & set2)
print('Difference :', set1 - set2)
print('Difference :', set2 - set1)
print('symmetric_Difference :', set2 ^ set1)


print('-'*50)

## clearing
print(my_set)
my_set.clear()
print(my_set)


print('-'*50)
```

# Tuples

```python
## Creating a Tuples
tup = (1,2,3,4,5)
print(tup)
print(type(tup))
print('-'*30)


## Accessing alement
tup = (1,2,3,4,5)
print(tup[0])
print('-'*30)

## Slicing
tup = (1,2,3,4,5)
print(tup[0:3])
print('-'*30)

## Concatenate
tup1 = (1,2,3,4)
tup2 = (4,5,6,7)
tpl = tup1 + tup2
print(tpl)
print('-'*30)

## find lenght
print(len(tpl))
print('-'*30)

## Iterating
print(tpl)
for i in tpl:
    print(i)
print('-'*30)

## checking the element
print(tpl)
print(2 in tpl)
print(5 in tpl)
print(8 in tpl)

print('-'*30)


## checking the count
print(tpl)
print(tpl.count(2))
print('-'*30)

## finding the index
print(tpl)
print(tpl.index(2))
print('-'*30)

## multiplying the tuple
print(tpl)
print(tpl*3)
print('-'*30)
```

```python
## Problem Statement 2 : Longest Substring Without Repeating Characters

def length_of_longest_substring(s):
    char_set = set()
    max_length = 0
    left_pointer = 0

    for right_pointer in range(len(s)):
        while s[right_pointer] in char_set:
            char_set.remove(s[left_pointer])
            left_pointer += 1

        char_set.add(s[right_pointer])
        max_length = max(max_length, right_pointer - left_pointer + 1)

    return max_length

input_str = "abcabcbb"
result = length_of_longest_substring(input_str)
print("Length of Longest Substring:", result)



```
