# String

```python
## concatenation
print("Divya " + "Prakash")
print('-'*20)
## string replication
print("Prakash " *3)
print('-'*20)

## string length
print(len("Divya Prakash"))
print('-'*20)

## string slicing
print("Divya Prakash"[0])
print("Divya Prakash"[12])
print("Divya Prakash"[-5])
print("Divya Prakash"[0:6])
print("Divya Prakash"[6:13])
print("Divya Prakash"[7:12])
print("Divya Prakash"[2:])
print("Divya Prakash"[:6])
print('-'*20)

## string cse conversion
print("Divya Prakash".lower())
print("Divya Prakash".upper())
print('-'*20)

## string stripping
print("       Praksh   ".strip())
print('-'*20)

## string replacing
print("Divya Prakash")
print("Divya Prakash".replace('h', '-'))
print("Divya Prakash".replace('ya','##'))
print('-'*20)

## string count
print("Divya Prakash".count('a'))
print("Divya Prakash".count('A'))
print("Divya Prakash".lower().count('a'))
print('-'*20)

## sting find
print("Divya Prakash".find('ash'))
print("Divya Prakash".count('pra'))
print('-'*20)

## string check
print('Prakash'.isalpha())
print('Prakash1'.isalpha())
print('1234'.isdigit())
print('1234a'.isdigit())
print('DIVYa'.isupper())
print('-'*20)


## string capitalization
print('divya Prakash'.capitalize())
print('divya prakash'.title())
print('-'*20)

## Check for star and end
print('Divya prakash'.startswith('Div'))
print('Divya prakash'.endswith('ash'))

print('-'*20)

## 
print('Divya Prakash'.center(20, "*"))
print('Divya Prakash'.ljust(20, "*"))
print('Divya Prakash'.rjust(20, "*"))
print('-'*20)




```
