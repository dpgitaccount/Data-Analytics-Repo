# 2. Arithematic operator-


```python
## Addition
print(3+4)
print(2.3+4.4)
print(3.4+4)
print(True + False)
print("Divya " + "Prakash")
print("------------")
## Substraction
print(3-4)
print(2.3-4.4)
print(3.4-4)
print(True - False)
print("------------")

## Multiplication
print(3*4)
print(2.3*4.4)
print(3.4*4)
print(True * False)
print("------------")

## Division
print(3/4)
print(2.3/4.4)
print(3.4/4)
print("------------")

## floor Division
print(4 // 3)
print(5.4 // 4.4)
print(3.4 // 4)
print("------------")

## Modulo
print(3%4)
print(2.3%.4)
print(3.4%4)
print("------------")

## Exponentiation
print(3 ** 4)
print(2.3 ** 4.4)
print(3.4 ** 4)
print("------------")

## abs
print(abs(34))
print(abs(-23))
print(abs(3.44))
print("------------")

## round
print(round(3.44445,3))
print(round(2.34,2))
print(round(-554,0))
print("------------")
## pow
print(pow(3,4))
print(pow(2.3,3.4))

```

# 1. Comparision operator

```python
## ==
print(1==2)
print(2==2)
print(1.0==1)
print("prakash"=="prakash")
print(True==False)
print(True==True)
print(False ==0)
print("-------------")
## !=
print(1!=2)
print(2!=2)
print(1.0!=1)
print(True!=False)
print(True!=True)
print(False !=0)
print("-------------")
## >
print(1>2)
print(2>2)
print(1.0>1)
print(True>False)
print(True>True)
print(False >0)
print("-------------")
## <
print(1<2)
print(2<2)
print(1.0<1)
print(True<False)
print(True<True)
print(False <5)
print("-------------")
## <=
print(1<=2)
print(2<=2)
print(1.0<=1)
print(True<=False)
print(True<=True)
print(False <=0)
print("-------------")
## >=
print(1>=2)
print(2>=2)
print(1.0>=1)
print(True>=False)
print(True>=True)
print(False >=0)
```

# 3. Logical Operator

```python
## AND
print(True and True )
print(True and False)
print(False and False)
print(False and True)
print("--------------")

## OR
print(True or True )
print(True or False)
print(False or False)
print(False or True)
print("--------------")

## NOT
print(not True )
print(not False)
```
