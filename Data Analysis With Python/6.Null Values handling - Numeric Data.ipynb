# 2nd Method - Filling Value Instead Of Null Values

```python
## Usnig Imputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') 

## Imputer Transform Process
imputer.fit(df.iloc[:, 1:3].values)
x = imputer.transform(df.iloc[:, 1:3].values)

## original value after filling nan
df.iloc[:, 1:3] = x
df

## Converting in Integer Value
data = {'Country': ['France', 'Spain', np.nan, 'Spain', 'Germany', 'France', 'Spain', 'France', 'Germany', 'France'],
        'Age': [44.000000, 27.000000, 30.000000, 38.000000, 40.000000, 35.000000, 38.777778, 48.000000, 50.000000, 37.000000],
        'Salary': [72000.000000, 48000.000000, 54000.000000, 61000.000000, 63777.777778, 58000.000000, 52000.000000, 79000.000000, 83000.000000, 67000.000000],
        'Purchased': ['Yes', 'Yes', np.nan, 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', np.nan]}

# Create a DataFrame
df = pd.DataFrame(data)

# Round the 'Age' and 'Salary' columns
df['Age'] = df['Age'].round(0).astype('Int64')  
df['Salary'] = df['Salary'].round(2)
print(df)

```
