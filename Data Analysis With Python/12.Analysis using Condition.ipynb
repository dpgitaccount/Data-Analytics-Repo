# Conditional Analysis

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
df.head(5)
```

1.How many free apps are there in ART_AND_DESIGN?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
print(len(df_pr[df_pr['Type'] == 'Free']))
```

2. How many apps are there in ART_AND_DESIGN with rating more then 4.5?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
print(len(df_pr[df_pr['Rating'] > 4.5]))
```

3. How many apps are there in FAMILY with rating more then 4.5 and Free?

```python
df_pr = df[df['Category'] == 'FAMILY']
df_pr = df_pr[df_pr['Rating'] > 4.5]
print(len(df_pr[df_pr['Type'] == 'Free']))
```

4. List all the free apps with rating more then 4.5 and category is FAMILY?

```python
df_pr = df[df['Category'] == 'FAMILY']
df_pr = df_pr[df_pr['Rating'] > 4.5]
result = df_pr[df_pr['Type'] == 'Free'].values
result
for i in result:
    print(i[0])
```
