# Group by Function

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")

impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute.fit(df.iloc[ : , 2:3]. values)
df.iloc[ : , 2:3 ] = impute.transform(df.iloc[ :, 2:3]. values)
df = df.dropna()
```

```python
df.head()
```

1. Name the apps that are in ART_AND_DESIGN with rating more then 4.5 in ascending order WRT Rating?


```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values(by = 'Rating', ascending = True)#.head()
```

2. which category having maximum average rating

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Verify the data type of the 'Rating' column
print(df['Rating'].dtype)  # It should be float64 after conversion

# Drop rows where 'Rating' is NaN
df_clean = df.dropna(subset=['Rating'])

# Perform the groupby operation and calculate the mean on numeric values
result = df_clean.groupby('Category')['Rating'].mean().sort_values(ascending = False)

print(result)
```

3.How many free apps are there in each category in descending order?

```python
df_pr = df[df['Type'] == 'Free']
df_pr.groupby('Category').count()['Type'].sort_values(ascending = False)
```
